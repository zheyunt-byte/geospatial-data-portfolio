**Task 13):** As we learned in lecture, the **humidex** is used in Canada to quantify our perception of how hot it feels under conditions of high humidity (i.e., a ‘real-feel’ temperature that factors in the humidity). To calculate the **humidex** for a day, it would be more appropriate to use the maximum temperature from the day, rather than the daily mean temperature since the afternoons tend to be warmer. $$  \text{humidex} = T_{max} + \frac{5}{9} (e*10-10) $$

where 
- $\text{humidex}$ is the humidex value (no unit) 
- $T_{max}$ is the daily maximum air temperature in **Celsius (C)**
- $e$ is the vapour pressure in **kPa**
